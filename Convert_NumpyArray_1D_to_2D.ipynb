# Convert One-Dimensional Numpy Arrays to Two-Dimensional Row or Column Matrix

A common reshaping pattern is the conversion of a one-dimensional array into a two-dimensional row or column matrix. You can do this with the reshape method, or more easily by making use of the newaxis keyword within a slice operation:

```python
import numpy as np
```

```python
x = np.array([1, 2, 3])
```

**Row vector via reshape**

```python
x.reshape((1, 3))
```

**Row vector via newaxis**

```python
x[np.newaxis, :]
```

```python
x[np.newaxis, ...]
```

**Column vector via reshape**

```python
x.reshape((3, 1))
```

**Column vector via newaxis**

```python
x[:, np.newaxis]
```

```python
x[..., np.newaxis]
```
